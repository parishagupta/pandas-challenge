```python
# Import pandas and pathlib modules
import pandas as pd
from pathlib import Path

# Locate and define csv files
schools_complete_csv = Path("Resources/schools_complete.csv")
students_complete_csv = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
schools_df = pd.read_csv(schools_complete_csv)
students_df = pd.read_csv(students_complete_csv)
schools_df
students_df
# Combine all data into a single dataset
schools_students_df = pd.merge(students_df, schools_df, how="left", on=["school_name", "school_name"])
schools_students_df.head()
```

```python
'DISTRICT SUMMARY
```

```python
# Calculate total number of schools
school_count = len(pd.unique(schools_students_df['school_name']))
print(f'Number of Unique Schools : {school_count}')
```

```python
# Calculate total number of students
student_count = len(schools_students_df['student_name'])
print(f'Number of Students : {student_count:5,.0f}')
```

```python
# Calculate total budget
budget_total=schools_df['budget'].sum()
print(f'Total Budget : ${budget_total:10,.0f}')
```

```python
# Calculate budget per student
budget_per_student=budget_total/student_count
print(f'Per Student Budget : ${budget_per_student:3,.2f}')
```

```python
# Calculate average math score
math_score_avg=students_df['math_score'].mean()
print(f'Average Math Score : {math_score_avg:2,.2f}')
```

```python
# Calculate average reading score
reading_score_avg=students_df['reading_score'].mean()
print(f'Average Reading Score : {reading_score_avg:2,.2f}')
```

```python
# Calculate % passing math (the percentage of students who scored >= 70)
math_score_passed=schools_students_df[(schools_students_df["math_score"] >= 70)].count()["student_name"]
math_passed_percent = math_score_passed / float(student_count) * 100
print(f'Percent of Students Passing Math : {math_passed_percent:2,.2F}%')
```

```python
# Calculate % passing reading (the percentage of students who scored >= 70)
reading_score_passed=schools_students_df[(schools_students_df["reading_score"] >= 70)].count()["student_name"]
reading_passed_percent = reading_score_passed / float(student_count) * 100
print(f'Percent of Students Passing Reading : {reading_passed_percent:2,.2F}%')
```

```python
# Calculate % passing in all (the percentage of students who scored >= 70 overall)
combo_score_passed=schools_students_df[(schools_students_df["math_score"] >= 70)&(schools_students_df["reading_score"] >= 70)].count()["student_name"]
combo_passed_percent = combo_score_passed / float(student_count) * 100
print(f'Percent of Students Passing Overall : {combo_passed_percent:2,.2F}%')
```

```python
# Create a high-level snapshot of the district's calculated key metrics in a DataFrame
district_summary_df=pd.DataFrame({"Total Number of Schools":school_count,
                                  "Total Students": student_count,
                                  "Total Budget": budget_total,
                                  "Avg Math Score":math_score_avg,
                                  "Avg Reading Score":reading_score_avg,
                                  "% Passing Math":math_passed_percent,
                                  "% Passing Reading":reading_passed_percent,
                                  "% Overall Passing":combo_passed_percent},index=[0])

# Format values appropriately
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.0f}".format)
district_summary_df["Avg Math Score"] = district_summary_df["Avg Math Score"].map("{:,.2f}%".format)
district_summary_df["Avg Reading Score"] = district_summary_df["Avg Reading Score"].map("{:,.2f}%".format)
district_summary_df["% Passing Math"] = district_summary_df["% Passing Math"].map("{:,.2f}%".format)
district_summary_df["% Passing Reading"] = district_summary_df["% Passing Reading"].map("{:,.2f}%".format)
district_summary_df["% Overall Passing"] = district_summary_df["% Overall Passing"].map("{:,.2f}%".format)

# Display the District Summary DataFrame
district_summary_df.style.hide()
```

```python
'SCHOOL SUMMARY
```

```python
# Store type per school
school_types = schools_df.set_index("school_name")["type"]
school_types
```

```python
# Calculate students per school
students_per_school=students_df["school_name"].value_counts()
students_per_school
```

```python
# Calculate total budget per school
budget_per_school=schools_df.groupby(["school_name"])["budget"].sum()
budget_per_school
```

```python
# Calculate budget per student per school
budget_per_student=budget_per_school/students_per_school
budget_per_student
```

```python
# Calculate average math score per school
math_score_avg_school=students_df.groupby(["school_name"])["math_score"].mean()
math_score_avg_school
```

```python
# Calculate average reading score per school
reading_score_avg_school=students_df.groupby(["school_name"])["reading_score"].mean()
reading_score_avg_school
```

```python
# Calculate # students per school passing math (the percentage of students who scored >= 70)
students_passing_math = schools_students_df[(schools_students_df["math_score"] >= 70)]
students_passing_math_school = students_passing_math.groupby(["school_name"]).size()
students_passing_math_school
```

```python
# Calculate # students per schools passing reading (the percentage of students who scored >= 70)
students_passing_reading = schools_students_df[(schools_students_df["reading_score"] >= 70)]
students_passing_reading_school = students_passing_reading.groupby(["school_name"]).size()
students_passing_reading_school
```

```python
# Calculate # students per schools passing math and reading (the percentage of students who scored >= 70 in both)
students_passing_math_and_reading = schools_students_df[(schools_students_df["reading_score"] >= 70) & (schools_students_df["math_score"] >= 70)]
students_passing_math_and_reading_school = students_passing_math_and_reading.groupby(["school_name"]).size()
students_passing_math_and_reading_school
```

```python
# Calculate % sudents per school passing math and reading
math_passed_percent_school = students_passing_math_school / students_per_school
reading_passed_percent_school = students_passing_reading_school / students_per_school
combo_passed_percent_school = students_passing_math_and_reading_school / students_per_school * 100

combo_passed_percent_school
```

```python
# Create a high-level snapshot of each school's calculated key metrics in a DataFrame
school_summary_df=pd.DataFrame({"School Type":school_types,
                                "Total Students":students_per_school,
                                "Total Budget":budget_per_school,
                                "Per Student Budget":budget_per_student,
                                "Avg Math Score":math_score_avg_school,
                                "Avg Reading Score":reading_score_avg_school,
                                "Passing Math":math_passed_percent_school,
                                "Passing Reading":reading_passed_percent_school,
                                "Overall Passing":combo_passed_percent_school})

# Format values appropriately
school_summary_df["Total Students"] = school_summary_df["Total Students"].map("{:,}".format)
school_summary_df["Total Budget"] = school_summary_df["Total Budget"].map("${:,.0f}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,.0f}".format)
school_summary_df["Avg Math Score"] = school_summary_df["Avg Math Score"].map("{:,.2f}".format)
school_summary_df["Avg Reading Score"] = school_summary_df["Avg Reading Score"].map("{:,.2f}".format)
school_summary_df["Passing Math"] = school_summary_df["Passing Math"].map("{:,.2f}%".format)
school_summary_df["Passing Reading"] = school_summary_df["Passing Reading"].map("{:,.2f}%".format)
school_summary_df["Overall Passing"] = school_summary_df["Overall Passing"].map("{:,.2f}%".format)

# Display the School Summary DataFrame
school_summary_df.head()
```

```python
# Create a DataFrame that lists the average math score for students of each grade (9th, 10th, 11th, 12th) at each school.

math_9th=students_df.loc[students_df["grade"] == "9th"].groupby(["school_name"])['math_score'].mean()
math_10th=students_df.loc[students_df["grade"] == "10th"].groupby(["school_name"])['math_score'].mean()
math_11th=students_df.loc[students_df["grade"] == "11th"].groupby(["school_name"])["math_score"].mean()
math_12th=students_df.loc[students_df["grade"] == "12th"].groupby(["school_name"])["math_score"].mean()

#create a DF for Math Scores by Grade
math_by_grade_df=pd.DataFrame({"9th":math_9th, "10th": math_10th, "11th":math_11th, "12th":math_12th })

# Format values
math_by_grade_df["9th"] = math_by_grade_df["9th"].map("{:,.2f}".format)
math_by_grade_df["10th"] = math_by_grade_df["10th"].map("{:,.2f}".format)
math_by_grade_df["11th"] = math_by_grade_df["11th"].map("{:,.2f}".format)
math_by_grade_df["12th"] = math_by_grade_df["12th"].map("{:,.2f}".format)

math_by_grade_df.head()
```

```python
'READING SCORES BY GRADE
```

```python
# Create a DataFrame that lists the average reading score for students of each grade (9th, 10th, 11th, 12th) at each school.
reading_9th=students_df.loc[students_df["grade"] == "9th"].groupby(["school_name"])['reading_score'].mean()
reading_10th=students_df.loc[students_df["grade"] == "10th"].groupby(["school_name"])['reading_score'].mean()
reading_11th=students_df.loc[students_df["grade"] == "11th"].groupby(["school_name"])["reading_score"].mean()
reading_12th=students_df.loc[students_df["grade"] == "12th"].groupby(["school_name"])["reading_score"].mean()

#create a DF for reading Scores by Grade
reading_by_grade_df=pd.DataFrame({"9th":reading_9th, "10th": reading_10th, "11th":reading_11th, "12th":reading_12th })

# Format values
reading_by_grade_df["9th"] = reading_by_grade_df["9th"].map("{:,.2f}".format)
reading_by_grade_df["10th"] = reading_by_grade_df["10th"].map("{:,.2f}".format)
reading_by_grade_df["11th"] = reading_by_grade_df["11th"].map("{:,.2f}".format)
reading_by_grade_df["12th"] = reading_by_grade_df["12th"].map("{:,.2f}".format)

reading_by_grade_df.head()
```

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["< $585", "$585-630", "$630-645", "$645-680"]

# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = school_summary_df.copy()

# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(budget_per_student, spending_bins, labels=labels, right=False)
school_spending_df
```

```python
# Add column to original df of budget per student in bins
size_bins = [0, 1000, 2000, 5000]
labels_size = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_size_df=school_summary_df.copy()

school_size_df["School Size"]=pd.cut(budget_per_student,size_bins,labels=labels_size)
school_size_df
```
